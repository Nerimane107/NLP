```python
def tokenize():
    tokenizedTweets = []
    tweet_list = tweets_data[' TWEET'].tolist()


    for element in tweet_list:

        wordsArray = nltk.tokenize.wordpunct_tokenize(str(element))
        tweet =[]
        for token in wordsArray:
            tweet.append(token)
        tokenizedTweets.append(tweet)
    return tokenizedTweets
```

```python
def lemmatizatization(tokenizedTweets):
    lemmer = qalsadi.lemmatizer.Lemmatizer()
    lemmatized_list = []

    for element in tokenizedTweets :
        tweet =[]
        for token in element:
            tweet.append(lemmer.lemmatize(str(token)))
        lemmatized_list.append(tweet)
    return lemmatized_list
```

```python

def stopWords():
# Download the stop words for Arabic language
    nltk.download('stopwords')

# Import the stop words for Arabic language
    from nltk.corpus import stopwords
    stop_words_arabic = stopwords.words('arabic')

# Print the list of stop words
    return stop_words_arabic
```

```python
def replace_elongated(doc: str) -> str:
        p = re.compile(r"([^0-9])\1{2,}")

        r = re.findall(p, doc)
        doc = doc.lower()
        strList = doc.split()

        c = ""
        for s in strList:
            unelong = s
            m = re.search(p, s)
            if m:
                extra = m.group(0)
                if (len(s) - len(extra)) == 0:
                    # All Letters are the same
                    unelong = extra[0:3]
                else:
                    unelong = s.replace(extra, extra[0])
            c = c + unelong + " "

        doc = c.strip()

        return doc
```

```python
 def clean_str(doc: str) -> str:
        repl_dict = {
            "Ø£": "Ø§",
            "Ø¥": "Ø§",
            "Ø¢": "Ø§",
            "Ø©": "Ù‡",
            "_": " ",
            "-": " ",
            "/": " ",
            '"': " ",
            "Ù€": " ",
            "'": " ",
            "Ù‰": "ÙŠ",
            "\\": " ",
            "\n": " ",
            "\t": " ",
            "&quot;": " ",
            "?": " ? ",
            "ØŸ": " ? ",
            "Â¿": " Â¿ ",
            "!": " ! ",
            "@": " ",
            "^": " ",
            "\r": "\r",
            "ÙˆÙˆ": "Ùˆ",
            "ÙŠÙŠ": "ÙŠ",
            "Ø§Ø§": "Ø§",
        }

        for key in repl_dict.keys():
            if key in doc:
                doc = doc.replace(key, repl_dict[key])
        doc = doc.strip()

        return doc
    
```

```python
def remove_tashkeel(doc: str) -> str:
        # Removes Tashkeel from input text
        p_tashkeel = re.compile(
            r"[\u0616-\u061A\u064B-\u0652\u06D6-\u06ED\u08F0-\u08F3\uFC5E-\uFC63\u0670]"
        )

        doc = re.sub(p_tashkeel, "", doc)
        return doc
```

```python
def remove_emojis(doc: str) -> str:
    doc=emoji.demojize(doc, delimiters=(" ", " "))
    return doc
```

```python
def replace_emojis(elongatedFree_list):
    emojiFree_list = []
    for w in elongatedFree_list :
        emojiFree_sentence=[]
        for word in w:

            emojiFree_sentence.append(remove_emojis(word))
            print(emojiFree_sentence)
        emojiFree_list.append(emojiFree_sentence)
    return emojiFree_list
```

```python
def elongatedFree(tashkeelFree_list):
    elongatedFree_list = []
    for w in tashkeelFree_list:
        clean_sentence=[]
        for word in w:
            clean_sentence.append(replace_elongated(word))
        elongatedFree_list.append(clean_sentence)
    return elongatedFree_list
```

```python
def punctuationFree(elongatedFree_list):
    punctuations="?:!.,;"
    for w in elongatedFree_list:
        for word in w:
            if word in punctuations:
                w.remove(word)
    return elongatedFree_list
```

```python
def tashkeelFree(clean_list):
    tashkeelFree_list = []
    for w in clean_list:
        clean_sentence=[]
        for word in w:
            clean_sentence.append(remove_tashkeel(word))
        tashkeelFree_list.append(clean_sentence)
    return tashkeelFree_list
```

```python
def clean(filtered_list):
    clean_list = []
    for w in filtered_list:
        clean_sentence=[]
        for word in w:
            clean_sentence.append(clean_str(word))
        clean_list.append(clean_sentence)
    return clean_list    
```

```python
def stopWordsFree(tokenizedTweets):
    nltk.download('stopwords')

# Import the stop words for Arabic language
    from nltk.corpus import stopwords
    stop_words_arabic = stopwords.words('arabic')

# Print the list of stop words
    filtered_list = []
    for w in tokenizedTweets:
        filtered_sentence=[]
        for word in w:
            if word not in stop_words_arabic:
                filtered_sentence.append(word)
        filtered_list.append(filtered_sentence)
    return filtered_list
```

```python

def preprocess(flag) :
    
    tokenizedList=tokenize()
    stopWordsFreeList = stopWordsFree(tokenizedList)
    cleanList= clean(stopWordsFreeList)
    tashkeelFreeList = tashkeelFree(cleanList)
    elongatedFreeList =  elongatedFree(tashkeelFreeList)
    if(flag == True):
        lemmatizedList= lemmatizatization(elongatedFreeList)
        emojiFreeList = replace_emojis(lemmatizedList)
        punctFreeList=  punctuationFree(emojiFreeList)
        return punctFreeList
    else:
        emojiFreeList = replace_emojis(elongatedFreeList)
        punctFreeList=  punctuationFree(emojiFreeList)
        return punctFreeList
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import pytorch_lightning as pl
import pandas as pd
from transformers import AutoTokenizer, AutoModel
from torchmetrics import Accuracy
from pytorch_lightning.callbacks import ModelCheckpoint
from transformers import AutoTokenizer, AutoModel
tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabert")

class SentimentAnalysisModel(nn.Module):
    def __init__(self, num_labels):
        super(SentimentAnalysisModel, self).__init__()
        self.num_labels = num_labels
        self.bert = AutoModel.from_pretrained("aubmindlab/bert-base-arabertv02")
        self.dropout = nn.Dropout(0.3)
        self.linear = nn.Linear(768, num_labels)

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        # output = self.dropout(pooled_output)
        print(outputs[1].shape)
        logits = self.linear(outputs[1])
        return logits

class SentimentAnalysisModule(pl.LightningModule):
    def __init__(self, model, learning_rate=1e-4):
        super(SentimentAnalysisModule, self).__init__()
        self.model = model
        self.learning_rate = learning_rate
        self.loss_fn = nn.CrossEntropyLoss()
        self.accuracy = Accuracy(task="multiclass", num_classes=8)

    def forward(self, input_ids, attention_mask):
        return self.model(input_ids, attention_mask)

    def training_step(self, batch, batch_idx):
        input_ids, attention_mask, labels = batch
        logits = self.forward(input_ids, attention_mask)
        loss = self.loss_fn(logits, labels)
        self.log('train_loss', loss)
        acc = self.accuracy(logits, labels) 
        self.log('train_accuracy', acc, prog_bar=True)
        return loss

    def validation_step(self, batch, batch_idx):
        input_ids, attention_mask, labels = batch
        logits = self.forward(input_ids, attention_mask)
        loss = self.loss_fn(logits, labels)
        self.log('val_loss', loss)
        acc = self.accuracy(logits, labels) 
        self.log('val_accuracy', acc, prog_bar=True)

    def test_step(self, batch, batch_idx):
        input_ids, attention_mask, labels = batch
        logits = self.forward(input_ids, attention_mask)
        loss = self.loss_fn(logits, labels)
        self.log('test_loss', loss)

    def configure_optimizers(self):
        optimizer = optim.AdamW(self.model.parameters(), lr=self.learning_rate,weight_decay=0.01)
        return optimizer

class SentimentAnalysisDataModule(pl.LightningDataModule):
    def __init__(self, train_csv, val_csv,batch_size=32):
        super(SentimentAnalysisDataModule, self).__init__()
        self.train_csv = train_csv
        self.val_csv = val_csv
        self.batch_size = batch_size
        self.tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabertv02")
    def prepare_data(self):
        pass

    def setup(self, stage=None):
        if stage == 'fit' or stage is None:
            self.train_data = pd.read_csv(self.train_csv)
            self.val_data = pd.read_csv(self.val_csv)

    def __tokenize_data(self, data):
        # data = data.dropna()
#         data['LABEL'] = pd.Categorical(df2['LABEL']).codes
        encoded_data = self.tokenizer.batch_encode_plus(
            data['Tweet'].tolist(),
            padding='max_length',
            truncation=True,
            max_length=128,
            return_tensors='pt'
        )
        input_ids = encoded_data['input_ids']
        attention_mask = encoded_data['attention_mask']
        labels = torch.tensor(data['Label'].tolist())
        return input_ids, attention_mask, labels

    def train_dataloader(self):
        input_ids, attention_mask, labels = self.__tokenize_data(self.train_data)
        dataset = torch.utils.data.TensorDataset(input_ids, attention_mask, labels)
        return torch.utils.data.DataLoader(dataset, batch_size=self.batch_size, shuffle=True)

    def val_dataloader(self):
        input_ids, attention_mask, labels = self.__tokenize_data(self.val_data)
        dataset = torch.utils.data.TensorDataset(input_ids, attention_mask, labels)
        return torch.utils.data.DataLoader(dataset, batch_size=self.batch_size)

    def test_dataloader(self):
        input_ids, attention_mask, labels = self.__tokenize_data(self.test_data)
        dataset = torch.utils.data.TensorDataset(input_ids, attention_mask, labels)
        return torch.utils.data.DataLoader(dataset, batch_size=self.batch_size)

def inference(model, tokenizer, sentence):
    encoded_sentence = tokenizer.encode_plus(
        sentence,
        padding='max_length',
        truncation=True,
        max_length=128,
        return_tensors='pt'
    )
    input_ids = encoded_sentence['input_ids']
    attention_mask = encoded_sentence['attention_mask']
    logits = model(input_ids, attention_mask)
    probabilities = torch.softmax(logits, dim=1)
    probabilities = probabilities.squeeze().tolist()
    sentiment_labels = { 0: "none" , 1:"anger", 2:"sadness", 3: "joy", 4: "surprise", 5:"sympathy", 6:"fear", 7:"love" }
    predicted_labels = torch.argmax(logits, dim=1)
    print(predicted_labels.shape)
    sorted_probabilities = sorted(zip(sentiment_labels, probabilities), key=lambda x: x[1], reverse=True)
    mapped_list = [(sentiment_labels[pair[0]], pair[1]) for pair in sorted_probabilities]
    return sentiment_labels[predicted_labels.item()],mapped_list

# def main():
#     train_csv = 'train.csv'
#     val_csv = 'val.csv'
#     test_csv = 'test.csv'
#     batch_size = 32
#     learning_rate = 1e-4
#     num_labels = 8

#     data_module = SentimentAnalysisDataModule(train_csv, val_csv, test_csv, batch_size)
#     data_module.setup()

#     model = SentimentAnalysisModel(num_labels)
#     sentiment_module = SentimentAnalysisModule(model, learning_rate)

#     trainer = pl.Trainer(gpus=1, max_epochs=5)
#     trainer.fit(sentiment_module, data_module)

#     # Example inference
#     sentence = "This movie made me really happy!"
#     tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabertv02")
#     sorted_probabilities = inference(model, tokenizer, sentence)
#     print(f"Sentence: {sentence}")
#     for sentiment, probability in sorted_probabilities:
#         print(f"Sentiment: {sentiment}, Probability: {probability}")

# if __name__ == '__main__':
#     main()
```
